# –î–æ–º–∞—à–Ω—î –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ —Ç–µ–º—É Spark SQL

–ó–∞–¥–∞—á—ñ –∑ –¥–æ–º–∞—à–Ω—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ SQL –ø–æ—Ç—Ä—ñ–±–Ω–æ —Ä–æ–∑–≤ º—è–∑–∞—Ç–∏ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é Spark SQL DataFrame API.

- –î–∞–º–ø–∏ —Ç–∞–±–ª–∏—Ü—å –∑–Ω–∞—Ö–æ–¥—è—Ç—å—Å—è –≤ –ø–∞–ø—Ü—ñ `data`. –î–∞—Ç–∞—Ñ—Ä–µ–π–º–∏ —Ç–∞–±–ª–∏—Ü—å –≤–∂–µ —Å—Ç–≤–æ—Ä–µ–Ω—ñ –≤ –∫–ª—ñ—Ç–∏–Ω—Ü—ñ –≤–∏—â–µ.
- –ú–æ–∂–µ—Ç–µ —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ —Å—Ç—ñ–ª—å–∫–∏ –Ω–æ–≤–∏—Ö –∫–ª—ñ—Ç–∏–Ω–æ–∫, —Å–∫—ñ–ª—å–∫–∏ –≤–∞–º –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ.
- –†–æ–∑–≤ º—è–∑–æ–∫ –∫–æ–∂–Ω–æ—ó –∑–∞–¥–∞—á—ñ –º–∞—î –±—É—Ç–∏ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ —Å–∞–º–æ–º—É —Ñ–∞–π–ª—ñ (–≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ –º–µ—Ç–æ–¥ `.show()`)
- –∫–æ–¥ –º–∞—î –±—É—Ç–∏ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏–π —É –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ—Å—Ç—ñ —ñ–∑ –æ–¥–Ω–∏–º —ñ–∑ —Å—Ç–∏–ª–µ–º, –ø–æ–∫–∞–∑–∞–Ω–∏–º –ª–µ–∫—Ç–æ—Ä–æ–º –Ω–∞ –∑–∞–Ω—è—Ç—Ç—ñ 13.

**–£–≤–∞–≥–∞!**
–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –º–æ–≤—É –∑–∞–ø–∏—Ç—ñ–≤ SQL –±–µ–∑–ø–æ—Å–µ—Ä–µ–¥–Ω—å–æ –∑–∞–±–æ—Ä–æ–Ω—è—î—Ç—å—Å—è, –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –≤–∏–∫–ª—é—á–Ω–æ DataFrame API!


1.
–í–∏–≤–µ—Å—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ñ—ñ–ª—å–º—ñ–≤ –≤ –∫–æ–∂–Ω—ñ–π –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó.
–†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ñ–¥—Å–æ—Ä—Ç—É–≤–∞—Ç–∏ –∑–∞ —Å–ø–∞–¥–∞–Ω–Ω—è–º.

```python
from pyspark.sql import functions as F

# –ü—Ä–∏—î–¥–Ω—É—î–º–æ film_category –¥–æ category, —â–æ–± –æ—Ç—Ä–∏–º–∞—Ç–∏ –Ω–∞–∑–≤–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π
films_by_category = film_category_df \
    .join(category_df, "category_id", "inner") \
    .groupBy("category_id", "name") \
    .agg(F.count("film_id").alias("–∫—ñ–ª—å–∫—ñ—Å—Ç—å_—Ñ—ñ–ª—å–º—ñ–≤")) \
    .select("name", "–∫—ñ–ª—å–∫—ñ—Å—Ç—å_—Ñ—ñ–ª—å–º—ñ–≤") \
    .orderBy(F.col("–∫—ñ–ª—å–∫—ñ—Å—Ç—å_—Ñ—ñ–ª—å–º—ñ–≤").desc())

films_by_category.show(truncate=False)
```

2. –í–∏–≤–µ—Å—Ç–∏ 10 –∞–∫—Ç–æ—Ä—ñ–≤, —á–∏—ó —Ñ—ñ–ª—å–º–∏ –±—Ä–∞–ª–∏ –Ω–∞ –ø—Ä–æ–∫–∞—Ç –Ω–∞–π–±—ñ–ª—å—à–µ.
–†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ñ–¥—Å–æ—Ä—Ç—É–≤–∞—Ç–∏ –∑–∞ —Å–ø–∞–¥–∞–Ω–Ω—è–º.

```python
top_actors_detailed = rental_df \
    .join(inventory_df, "inventory_id", "inner") \
    .join(film_df, "film_id", "inner") \
    .join(film_actor_df, "film_id", "inner") \
    .join(actor_df, "actor_id", "inner") \
    .groupBy("actor_id", "first_name", "last_name") \
    .agg(
        F.count("rental_id").alias("–∫—ñ–ª—å–∫—ñ—Å—Ç—å_–ø—Ä–æ–∫–∞—Ç—ñ–≤"),
        F.countDistinct("film_id").alias("–∫—ñ–ª—å–∫—ñ—Å—Ç—å_—Ä—ñ–∑–Ω–∏—Ö_—Ñ—ñ–ª—å–º—ñ–≤"),
        F.countDistinct("rental_id").alias("—É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö_–ø—Ä–æ–∫–∞—Ç—ñ–≤")
    ) \
    .select(
        F.concat("first_name", F.lit(" "), "last_name").alias("–∞–∫—Ç–æ—Ä"),
        "–∫—ñ–ª—å–∫—ñ—Å—Ç—å_–ø—Ä–æ–∫–∞—Ç—ñ–≤",
        "–∫—ñ–ª—å–∫—ñ—Å—Ç—å_—Ä—ñ–∑–Ω–∏—Ö_—Ñ—ñ–ª—å–º—ñ–≤"
    ) \
    .orderBy(F.col("–∫—ñ–ª—å–∫—ñ—Å—Ç—å_–ø—Ä–æ–∫–∞—Ç—ñ–≤").desc()) \
    .limit(10)

top_actors_detailed.show(truncate=False)
```

3.
–í–∏–≤–µ—Å—Ç–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è —Ñ—ñ–ª—å–º—ñ–≤, –Ω–∞ —è–∫—É –±—É–ª–æ –≤–∏—Ç—Ä–∞—á–µ–Ω–æ –Ω–∞–π–±—ñ–ª—å—à–µ –≥—Ä–æ—à–µ–π
–≤ –ø—Ä–æ–∫–∞—Ç—ñ

```python
categories_by_revenue = payment_df \
    .join(rental_df, "rental_id", "inner") \
    .join(inventory_df, "inventory_id", "inner") \
    .join(film_df, "film_id", "inner") \
    .join(film_category_df, "film_id", "inner") \
    .join(category_df, "category_id", "inner") \
    .groupBy("category_id", "name") \
    .agg(F.sum("amount").alias("–∑–∞–≥–∞–ª—å–Ω–∞_—Å—É–º–∞")) \
    .select("name", F.round("–∑–∞–≥–∞–ª—å–Ω–∞_—Å—É–º–∞", 2).alias("–∑–∞–≥–∞–ª—å–Ω–∞_—Å—É–º–∞")) \
    .orderBy(F.col("–∑–∞–≥–∞–ª—å–Ω–∞_—Å—É–º–∞").desc())

categories_by_revenue.show(truncate=False)
```

4.
–í–∏–≤–µ—Å—Ç–∏ –Ω–∞–∑–≤–∏ —Ñ—ñ–ª—å–º—ñ–≤, —è–∫–∏—Ö –Ω–µ –º–∞—î –≤ inventory.

```python
films_not_in_inventory_v2 = film_df \
    .join(inventory_df, "film_id", "left") \
    .filter(F.col("inventory_id").isNull()) \
    .select("film_id", "title", "release_year", "rating", "length") \
    .orderBy("title")

print(f"\n–ö—ñ–ª—å–∫—ñ—Å—Ç—å —Ñ—ñ–ª—å–º—ñ–≤: {films_not_in_inventory_v2.count()}")
films_not_in_inventory_v2.show(truncate=False)
```
